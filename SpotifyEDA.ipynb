<h3>Feature Engineering: Correlation Map</h3>

```python
corr_df=df_tracks.drop(["id","name","artists","id_artists","key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(16,7))
cm = sns.dark_palette("red", as_cmap=True)
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g", vmin=-1, vmax=1, center=0, cmap=cm, linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

<h3>Correlation between Loudness and Energy</h3>

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
sample_df
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "g").set(title = "Loudness vs Energy")
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "c").set(title = "Popularity vs Accousticness")
```

<h3>Songs per Year</h3>

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
year = df_tracks.dates.dt.year
```

```python
year = pd.DataFrame(year)
df_tracks['year'] = year
df_tracks = df_tracks[df_tracks["year"]<2023]
```

```python
sns.displot(year,discrete = True, aspect = 2, height = 5, kind = "hist").set(title = "Songs per Year")
```

<h3>Duration vs Years</h3>

```python
total = df_tracks.duration_s
fig, ax = plt.subplots(figsize = (18,7))
fig = sns.barplot(x = df_tracks["year"], y = total, ax = ax, errwidth = False).set(title = "Duration vs Years")
plt.xticks(rotation=90)
```

```python
total = df_tracks.duration_s
sns.set_style(style = "whitegrid")
fig, ax = plt.subplots(figsize = (10,5))
fig = sns.lineplot(x = df_tracks["year"], y = total, ax = ax, color="c").set(title = "Duration vs Years")
plt.xticks(rotation=60)
```
